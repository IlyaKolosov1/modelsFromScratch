# Linear Regression

## Let's break down the key components of Linear Regression:

1. **Weights** – Parameters that adjust to generalize patterns in the data.  
2. **Features** – The input data used to make predictions.  
3. **Bias** – Accounts for cases where the distribution does not pass through the origin.  
4. **Loss function** – A function that updates the weights to minimize prediction errors and improve accuracy.
